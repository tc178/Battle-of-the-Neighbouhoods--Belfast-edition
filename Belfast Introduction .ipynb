## Business Problem:

While this is no doubt an excellent time and opportunity to open a business to cater for the incoming tourists; the question remains, where should someone (say John) open a hotel to provide accommodation for these guests?
Belfast is made up of 18 Postcodes, all of which offer already existing different attractions/entertainment/restaurants. Therefore the best method for John to decide which area to open a hotel in, is to explore the locational data behind each postcode, to see what each neighbourhood has to offer. Using this data, John can then make a well informed decision, which should maximise the profit of this business venture. 


## The Data:

Now that John has the problem and the business opportunity outlined, how is he going to solve this using data? Fortunately Belfast has a wealth of business and corresponding locational data. Belfast is split up into 18 postcodes, with each district and location providing a range of restaurants/bars/entertainment. We can therefore use foursquares locational data, analyse the different Districts/Neighbourhoods to see which area offers the most/least competition to open a hotel; meanwhile analysing each neighbourhood to see what they offer to attract and entertain hotel guests in the nearby area.
